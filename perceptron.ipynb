# PERCEPTRON
## Perceptron algorithm finds a line which seperates linear seperable data in n-dimensions. Here an instance of perceptron is depicted where dimension is 1 and 2.(because more dimensions cannot be plotted). But nonetheless the algorithm works fine for n-dimensions, the only problem is with plotting.
## In each iteration of the algorithm the "so found" line is plotted using basic trick. Also the line vector is printed.
## Final line separates the linear seperable data completely with no error.
## Here linear seperable data in n-dimensions is also generated using concept of hyperplanes.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

#generating linear seperable data in dim-dimensions 
def gen_linear_seperable_data(dim, gamma, R, n):
    label_pos1_points=[]
    label_neg1_points=[]
    w=np.random.rand(dim)
    w_hat=w/(np.linalg.norm(w))
    while(True):
        x=np.random.uniform(-R, R, dim)
        if((np.dot(w_hat, x))>=gamma):
            label_pos1_points.append([x, +1])
            #label is +1 for these points
            if(len(label_pos1_points)==int(n/2)):
                break;
    while(True):
        x=np.random.uniform(-R, R, dim)
        if((np.dot(w_hat, x))<=-gamma):
            label_neg1_points.append([x, -1])
            #label is -1 for these points
            if(len(label_neg1_points)==int(n/2)):
                break
    # each category(+1, -1) data has same number of points
    return label_pos1_points, label_neg1_points

#plotting the created dataset to understand how the data is distributed
#blue color for +1 label points, red color for -1 label points
def dataset_plotter(dim, points):
    if(dim==1):
        for point in points:
            if(point[1]==+1):
                plt.scatter(point[0], 0, marker='.', color='blue')
            if(point[1]==-1):
                plt.scatter(point[0], 0, marker='.', color='red')
    elif(dim==2):
        for point in points:
            if(point[1]==+1):
                plt.scatter(point[0][0], point[0][1], marker='.', color='blue')
            if(point[1]==-1):
                plt.scatter(point[0][0], point[0][1], marker='.', color='red')

#perceptron algorithm for 1d and 2d
def perceptron(points, w, g, dim):
    itr=0
    w1=np.array([])
    if(dim==2):
        while(not np.array_equal(w1, w)):
            w1=np.copy(w)
            dataset_plotter(dim, points)
            plt.plot([-w[1], w[1]], [w[0], -w[0]], color='k')
            plt.show()
            print('w=', w)
            for point in points:
                x=np.array(point[0])
                if((point[1]*(np.dot(w, x)))<=0):
                    w=(point[1]*x)+w
                    itr+=1
                    break
    elif(dim==1):
        while(not np.array_equal(w1, w)):
            w1=np.copy(w)
            dataset_plotter(dim, points)
            plt.plot([0, 0], [-w[0], w[0]], color='k')
            plt.show()
            print('w=', w)
            for point in points:
                x=np.array(point[0])
                if((point[1]*(np.dot(w, x)))<=0):
                    w=(point[1]*x)+w
                    itr+=1
                    break
    print('iterations=', itr)
    return w
                
##############################################################################################################
#fixing parameters n-size of dataset, dim-dimension of dataset, gamma-distance between two labelled categories
#       of dataset, R-radius bound for dataset
n=500
dim=2
gamma=1
R=10
points1, points2=gen_linear_seperable_data(dim, gamma, R, n)
#appending two categories of dataset to make one dataset(here data is not shuffled but shuffling is 
#       advisable)
points=[*points1, *points2]
#putting initial w to be zero
w_initial=np.zeros(dim)
w_final=perceptron(points, w_initial, gamma, dim)
#note: for dim=1 line which seperates data is found in 1 iteration
```
